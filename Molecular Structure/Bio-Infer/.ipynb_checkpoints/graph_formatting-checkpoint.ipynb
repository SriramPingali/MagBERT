# NODE 

## Set Up

```python
import numpy as np
from tqdm import tqdm_notebook
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
with open('biomed.train', 'r') as f:
    lines = f.readlines()
```

## Vocabulary

```python
corpus = []
labels = []
proteins = []
for line in lines:
    temp = line.split('\t')
    proteins.append(temp[1:-1])
    corpus.append(temp[0].lower())
    labels.append(temp[-1].split('\n')[0])
```

```python
vocab = []
for doc in corpus:
    vocab += doc.split()

vocab = list(set(vocab))
values = np.array(vocab)
```

```python
values.size
```

```python
# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(values)
print(integer_encoded.size)
```

## Encode the Sentences / Documents

```python
def encode_sentence(sentence):
    indices = label_encoder.transform(sentence.split())
    vector = np.zeros(integer_encoded.size, dtype = 'int')
    for index in indices:
        vector[index] = 1
    return(vector)
```

```python
# index = 0
# nodes = []
enc_corpus = []
for doc, label in zip(corpus, labels):
#     temp = ''
#     temp += str(index) + '\t'
#     temp += '\t'.join([str(n) for n in encode_sentence(doc)])
#     temp += '\t'+ label + '\n'
#     index += 1
#     nodes.append(temp)
    enc_corpus.append(encode_sentence(doc))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=835, svd_solver='arpack')
red_dim_corp = pca.fit_transform(enc_corpus)
```

```python
red_dim_corp
```

```python
index = 0
nodes = []
for doc, label in zip(red_dim_corp, labels):
    temp = ''
    temp += str(index) + '\t'
    temp += '\t'.join([str(n) for n in doc])
    temp += '\t'+ label + '\n'
    index += 1
    nodes.append(temp)
```

```python
nodes[0]
```

```python
with open('node', 'w') as f:
    f.writelines(nodes)
```

# EDGES

```python
protein_corpus = []
for doc_proteins in proteins:
    for protein in doc_proteins:
        protein_corpus.append(protein)
        
protein_corpus = list(set(protein_corpus))
```

```python
protein_encoder = LabelEncoder()
protein_encodings = protein_encoder.fit_transform(protein_corpus)
```

```python
protein_encoder.transform(['skeletal actin'])
```

```python
doc_protein_encodings = []
for protein in proteins:  
    doc_protein_encodings.append(protein_encoder.transform(protein))
```

```python
doc_protein_encodings
```

```python
edges = []
for i in tqdm_notebook(range(len(doc_protein_encodings))):
    for j, compare_to in enumerate(doc_protein_encodings):
        for compare in doc_protein_encodings[i]:
            if i != j and compare in compare_to:
                edges.append(str(i) + "\t" + str(j) + "\n")
                break
```

```python
len(edges)
```

```python
with open('link', 'w') as f:
    f.writelines(edges)
```
